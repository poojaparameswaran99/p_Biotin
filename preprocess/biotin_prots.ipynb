```python
def get_PTMlocation(df, peptide_col: str, peptide_position_col: str, biotin_keyword='[Biotin]'):
    ## assumes one match 
    biotin_loc=df[peptide_col].str.strip('_').str.find(biotin_keyword) ## find requires exact match
    biotin_loc = biotin_loc.replace(-1, np.nan)
    df['BiotinPosition'] = df.loc[:, peptide_position_col].str.split(';|,|:').str[0].astype(int) + biotin_loc -1
    return df

def clean_peptide(df, peptide_col):
    df[peptide_col] = df[peptide_col].str.replace(r'\[(?!Biotin\])[^\]]*\]|_', '',regex=True)
    return df
```

```python
df1 = df1.pipe(clean_peptide, 'Sequence').pipe(get_PTMlocation, 'Sequence', 'PeptidePosition')
df2 = df2.pipe(clean_peptide, 'Sequence').pipe(get_PTMlocation, 'Sequence', 'PeptidePosition')
```

```python
##ID for each protein, bniotinylated sites
tot = pd.concat([df1[['Accession', 'BiotinPosition']].dropna(), df2[['Accession', 'BiotinPosition']].dropna()])
bsites = tot.groupby(['Accession'])['BiotinPosition'].agg(lambda x: sorted(list(set(x)))).reset_index()
bsites.to_csv('../data/Protein_w_BiotinSites.csv', index=False)
```
